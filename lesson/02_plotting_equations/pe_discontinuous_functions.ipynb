## 1. Heavyside Step Function

A function which takes the value 0 for values less than 0 and 1 otherwise.

$$H(x) = \left \{ \begin{matrix*}[r l] 1 & \text{ for } x \geq 0 \\
0 & \text{ for } x < 0 \end{matrix*} \right.$$

```python
import numpy as np
import matplotlib.pyplot as plt



x = np.linspace(-5,5,1000)
y = 0*x
z = 0*x + 1

zero_restriction = x<0
one_restriction = x>=0

plt.figure()
plt.plot(x[zero_restriction],y[zero_restriction], color = 'k')
plt.plot(x[one_restriction],z[one_restriction], color = 'k')
plt.plot([0],[1],'ko')
plt.plot([0],[0],'ko', fillstyle = 'none')
plt.axvline(0,color='k',linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("Heavyside Step Function")
plt.show()
```

## 2. The ratio between $x^3$ and $x^2 - 2 x + 1$.

Vertical Asymptote at x = 1

$$\frac{x^3}{(x-1)^2}$$


```python
x = np.linspace(-10,10,1000)
y = (x * x * x) / ((x-1) * (x-1))

plt.figure()
plt.plot(x,y, label = "$\\frac{x^3}{(x-1)^2}$", color = "k")
plt.axvline(0,color='k',linestyle='-')
plt.axvline(1,color='k',linestyle='--')
plt.axhline(0,color='k',linestyle='-')
plt.xlim([None,None])
plt.ylim([-50,50])
plt.legend() 
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("Plot of the ratio between $x^3$ and $x^2 - 2 x + 1$")
plt.show()
```

## 3. The ratio between $x^2 - 1$ and $x - 1$

$$\frac{x^2-1}{x-1}$$

$$\frac{(x+1)(x-1)}{x-1}$$

$$(x+1)$$


```python
x = np.linspace(-10,10,1000)
y = ((x * x) - 1) / (x-1)

plt.figure()
plt.plot(x,y, color = 'k', label = "$\\frac{(x^2-1)}{(x-1)}$")
plt.axvline(0,color='k',linestyle='-')
plt.axhline(0,color='k',linestyle='-')
plt.plot([1],[2],'ko', fillstyle = 'none')
plt.xlim([-5,5])
plt.ylim([-5,5])
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("Plot of the ratio between $x^2 - 1$ and $x - 1$")
plt.show()
```

## 4. The Tangent Function 

Vertical Asymptotes at $\frac{\pi}{2}$ intervals. 

$tan (x) $

```python
x = np.linspace(-np.pi,np.pi,1000)
y = np.tan(x)

pi_asymptote = np.pi/2

plt.figure()
plt.plot(x,y, marker = 'o', color = 'k', label = "$tan (x)$")
plt.axhline(0,color='k',linestyle='-')
plt.axvline(pi_asymptote, color = 'blue', linestyle = '--')
plt.axvline((pi_asymptote - np.pi), color = 'blue', linestyle = '--')
plt.xlim([-np.pi,np.pi])
plt.ylim([-10,10])
plt.legend()
plt.xlabel('x \n Note: Vertical Asymptotes plotted in blue')
plt.ylabel('f(x)')
plt.title("Plot of Tangent Function")
plt.show()
```
