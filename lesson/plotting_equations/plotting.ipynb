# Basic plotting

One of the Python packages which can be convenient for plotting things is `matplotlib.pyplot`. Originally, it was modeled on Matlab's plotting system.

Create a new figure with the `pyplot.figure()` function. Then use the `pyplot.plot()` function to plot two equal-length vectors on a two-dimensional plane.

`pyplot.show()` forces Python to display the figure.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1,5,1000)

y = 2*x**3 - 10*x**2 + 5*x - 2

plt.figure()
plt.plot(x,y)
plt.show()
```

We can plot some horizontal and vertical lines (such as the axes) by using `pyplot.axvline` and `pyplot.axhline`. `k` is shorthand for black color, and '--' denotes a dashed line.

```python
plt.figure()
plt.plot(x,y)
plt.axvline(0,color='k',linestyle='--')
plt.axhline(0,color='k',linestyle='--')
plt.show()
```

We can plot as many functions as we like on the same plot. If we don't specify the color, `pyplot` will automatically cycle through a standard color palette:

```python
x = np.linspace(-1,5,1000)

y = 2*x**3 - 10*x**2 + 5*x - 2
z = -0.5*x**3 + 3*x**2 - 10*x +1
q = 5*np.sin(5*x)

plt.figure()
plt.plot(x,y)
plt.plot(x,z)
plt.plot(x,q)
plt.axvline(0,color='k',linestyle='--')
plt.axhline(0,color='k',linestyle='--')
plt.show()
```

We can use `xlim` and `ylim` to limit what the plot window shows us. In the case of functions that tend towards infinity, this can be important!

```python
x = np.linspace(-1,5,1000)

y = 2*x**3 - 10*x**2 + 5*x - 2
z = -0.5*x**3 + 3*x**2 - 10*x +1
q = 5*np.sin(5*x)
r = np.tan(x)

plt.figure()
plt.plot(x,y)
plt.plot(x,z)
plt.plot(x,q)
plt.plot(x,r)
plt.axvline(0,color='k',linestyle='--')
plt.axhline(0,color='k',linestyle='--')
plt.title('without limiting plot window')
plt.show()

plt.figure()
plt.plot(x,y)
plt.plot(x,z)
plt.plot(x,q)
plt.plot(x,r)
plt.axvline(0,color='k',linestyle='--')
plt.axhline(0,color='k',linestyle='--')
plt.xlim([None,None])
plt.ylim([-25,25])
plt.title('with limiting plot window')
plt.show()
```

You can also label the axes, and make a legend by labeling any formulas you want included and then calling the `pyplot.legend()` function:

```python
plt.figure()
plt.plot(x,y,label='$f(x) = 2 x^3 - 10 x^2 + 5 x - 2$')
plt.plot(x,z,label='$f(x) = -0.5 x^3 + 3 x^2 - 10 x +1$')
plt.plot(x,q,label='$f(x) = 5 \cdot \sin ( 5 x )$')
plt.plot(x,r,label='$f(x) = \\tan x$')
plt.axvline(0,color='k',linestyle='--')
plt.axhline(0,color='k',linestyle='--')
plt.xlim([None,None])
plt.ylim([-25,25])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('example with 4 functions, axis labels, and a legend')
plt.show()
```
